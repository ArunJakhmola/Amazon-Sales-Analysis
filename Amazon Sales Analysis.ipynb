### There are two columns with zero values, we need to drop them. i.e. 'New' and 'PendingS'

```python
df.drop(['New', 'PendingS'], axis = 1, inplace = True)
```

```python
df.info()
```

## Check for the number of null values in the data

```python
df.isnull().sum()
```

There are more than 120,000 rows of data and we have more than 7000 null values we can not fill these many null cells with median or mean it can make huge difference in the final result, so removing null rows won't make much of a difference

```python
df = df.dropna(subset = ['currency', 'Amount'])
```

```python
df.isnull().sum()
```

```python
# Column '['fulfilled-by']' has a lot of null values which make it of no use for us. Hence we will delete the whole column.


df.drop(['fulfilled-by'], axis = 1, inplace = True)
```

```python
df.shape
```

## Data type change

```python
df.info()
```

Since there are only 35 null values in the following columns, which are not letting us move ahead with the analysis. So we will replace them with Mode of respective column.

```python
# df[['ship-city', 'ship-state', 'ship-postal-code', 'ship-country']] = df[['ship-city', 'ship-state', 'ship-postal-code', 'ship-country']].fillna(0)
df['ship-city'].fillna(df['ship-city'].mode()[0], inplace=True)
df['ship-state'].fillna(df['ship-state'].mode()[0], inplace=True)
df['ship-postal-code'].fillna(df['ship-postal-code'].mode()[0], inplace=True)
df['ship-country'].fillna(df['ship-country'].mode()[0], inplace=True)
```

```python
df['ship-postal-code'] = df['ship-postal-code'].astype('int')
```

```python
df.info()
```

```python
df.isna().sum()
#No null values left
```

```python
df.Date = pd.to_datetime(df.Date)
```

# Exploratory Data Anlaysis

```python
df.columns
```

### Size

```python
ax = sns.countplot(x = df.Size)

#To add the count values on the top of the bars
for bar in ax.containers:
    ax.bar_label(bar)
```

#### Note: From above Graph you can see that most of the people buy M-Size

## Group By

```python
g1 = df.groupby([df.Size], as_index = False)['Quantity'].sum()
```

```python
g1.sort_values(by = 'Quantity', ascending = False, inplace = True)
g1
```

```python
sns.barplot(x = g1.Size, y = g1.Quantity)
```

#### Note: From above graph you can see that M size has the most amount of sales, followed by L and XL.

## Courier Status

```python
plt.figure(figsize = (10, 6))
ax1 = sns.countplot(data = df, x = 'Courier Status', hue = 'Status')
plt.title('Courier Status')
plt.ylabel("Number of Orders")

plt.show()
```

#### Note: From our graph the majority of orders are shipped through the Courier.

```python
df.columns
```

```python
#histogram

df.Category = df.Category.astype(str)
category_data = df.Category
plt.figure(figsize = (10, 5))
plt.hist(category_data, bins = 18, edgecolor = 'black')
plt.show()
```

#### Note: From the above graph we can see that the T-shirts are sold the most.

```python
#Checking B2B data by using pie chart

b2b = df.B2B.value_counts()
labels = ['Other', 'B2B']
#plot the pie chart
plt.pie(b2b, labels = labels, autopct = '%1.1f%%')
plt.title('Distribution of B2B Sales')
plt.show()
```

#### Note: From the above chart we can see that maximum of the sales are from B2B channels i.e. 99 .3%

```python
#Prepare the data for scatter plot for Size

plt.figure(figsize = (10, 8))
plt.scatter(y = df.Size, x= df.Category)
plt.title('Scatter Plot')
plt.xlabel('Category')
plt.ylabel('Size')
plt.show()
```

```python
#Plotcount for cities by state

sns.set_style("whitegrid")
plt.figure(figsize = (15, 12))
sns.countplot(x = df['ship-state'], hue = df['ship-state'])
plt.xticks(rotation = 90)
plt.xlabel('State')
plt.ylabel('Count')
plt.show()
```

```python
#Top 10 states/cities

top_10_state = df['ship-state'].value_counts().head(10)

plt.figure(figsize = (12, 8))
sns.countplot(x='ship-state', data=df, order=top_10_state.index, palette="Set2")
plt.xticks(rotation = 90)
plt.xlabel('State')
plt.ylabel('Count')
plt.show()
```

#### Note: From the above graph we can see that most of the buyers are from Maharashtra

### Conclusion: The data analysis reveals that the business has a significant customer base in Maharashtra mainly serves retailers fulfills order through Amazon experiences high demand for T-shirts and the M size as a preferred choice among the buyers.
